```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
data=pd.read_csv("/content/airqualityindexdata/AQI Data Set.csv")

# Clean column names by removing leading/trailing spaces and replacing special characters/spaces with underscores
data.columns = data.columns.str.strip().str.replace('[^A-Za-z0-9_]+', '_', regex=True)

print("\nCleaned column names:", data.columns)

print("\nhead of the dataset")
display(data.head())
print("\nshape of the data set")
display(data.shape)
print("\ninfo of the dataset")
display(data.info())
print("\nsummary statistics of the dataset")
display(data.describe())
print("\nduplicated of the dataset")
display(data.duplicated())
print("\ncheck of missing data")
display(data.isnull().sum())
print("\nfill missing values with the mean of the each column")
# Calculate the mean for numeric columns only
numeric_cols = data.select_dtypes(include=np.number).columns
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())
print("\ncheck of missing data after filling")
display(data.isnull().sum())
print("\nshape of the dataset after filling")
display(data.shape)


def cap_outer(df, col):
  sorted_col = sorted(df[col])
  Q1, Q3 = pd.Series(sorted_col).quantile([0.25, 0.75])
  IQR = Q3 - Q1
  lower_bound = (Q1 - 1.5 * IQR)
  upper_bound = (Q3 + 1.5 * IQR)
  df[col] = np.clip(df[col], lower_bound, upper_bound)
  return df

numeric_cols_for_capping = data.select_dtypes(include=np.number).columns
for col in numeric_cols_for_capping:
  data = cap_outer(data, col)

print("\nshape of the dataset after outlier capping")
display(data.shape)

data.boxplot(column=["AQI"])
plt.show()


#nomalization and scaling

scaler=MinMaxScaler()
data_scaled= pd.DataFrame(scaler.fit_transform(data.select_dtypes(include=[np.number])),
                     columns=data.select_dtypes(include=[np.number]).columns)
display(data_scaled.head())
#encoding categorical variables
data_encoded=pd.get_dummies(data,drop_first=True)
#plot histograms for numerical columns
data.select_dtypes(include=[np.number]).hist(bins=15,figsize=(15,10))
plt.show()

#univariate analysis
plt.figure(figsize=(10,6))
sns.histplot(data['AQI'])
plt.title('distribution of AQI')
plt.show()
#Bivariate analysis
plt.figure(figsize=(12,8))
sns.heatmap(data.select_dtypes(include=np.number).corr(),annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title('correlation Heatmap')
plt.show()
sns.pairplot(data[['CO_in_mg_m3','NOx_in_g_m3','O3_in_g_m3','PM2_5_in_g_m3','AQI']])
plt.show()
 #correlation heatmap
plt.figure(figsize=(10,6))
sns.histplot(data['AQI'],kde=True)
plt.title('Distribution of AQI')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(x=data['AQI'])
plt.title('boxplot of AQI')
plt.show()
#boxpots to check distributions and outliers

for col in data.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(8,6))
    sns.boxplot(x=data[col])
    plt.title(f'boxplot of {col}')
    plt.show()
#save the cleaned dataset to a new csv file
data.to_csv('cleaned_AQI_data.csv',index=False)
```

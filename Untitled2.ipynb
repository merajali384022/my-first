```python
# Visualize the distribution of numerical columns for outlier detection
data.boxplot(column=['weekly_self_study_hours', 'attendance_percentage', 'class_participation', 'total_score'])
plt.show()
```

```python
def remove_outliers(col):
   sorted_col = sorted(col) # It's better to work with the original Series for quantile calculation
   Q1,Q3=col.quantile([0.25,0.75])
   IQR=Q3-Q1
   lower_range=Q1-(1.5*IQR)
   upper_range=Q3+(1.5*IQR)
   return lower_range,upper_range # Return the calculated ranges
```

```python
lrincome,urincome=remove_outliers(data['weekly_self_study_hours'])
data['weekly_self_study_hours']=np.where(data['weekly_self_study_hours']>urincome,urincome,data['weekly_self_study_hours'])
data['weekly_self_study_hours']=np.where(data['weekly_self_study_hours']<lrincome,lrincome,data['weekly_self_study_hours'])

lrincome,urincome=remove_outliers(data['attendance_percentage'])
data['attendance_percentage']=np.where(data['attendance_percentage']>urincome,urincome,data['attendance_percentage'])
data['attendance_percentage']=np.where(data['attendance_percentage']<lrincome,lrincome,data['attendance_percentage'])

lrincome,urincome=remove_outliers(data['class_participation'])
data['class_participation']=np.where(data['class_participation']>urincome,urincome,data['class_participation'])
data['class_participation']=np.where(data['class_participation']<lrincome,lrincome,data['class_participation'])

lrincome,urincome=remove_outliers(data['total_score'])
data['total_score']=np.where(data['total_score']>urincome,urincome,data['total_score'])
data['total_score']=np.where(data['total_score']<lrincome,lrincome,data['total_score'])

data.shape
```
